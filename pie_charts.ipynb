```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.font_manager as fm
from cycler import cycler
from matplotlib.font_manager import FontProperties
```

```python
from matplotlib.colors import LinearSegmentedColormap
colors_laura = ['#96C3DC','#1C63A5','#47ADB3','#A4DB77','#6DB74C','#117F01','#F78588','#D90016','#FFB578','#FD6906','#BEA1CC','#682294']
# Define the colorblind-friendly colormap
colors = ['#a6cee3', '#1f78b4', '#b2df8a', '#77BA99', '#fb9a99', '#e31a1c', '#fdbf6f', '#ff7f00', '#cab2d6', '#6a3d9a',
          '#ffff99', '#b15928', '#8dd3c7', '#5b9c0a', '#1b9e77', '#D4AFCD']

# Create the custom colormap
colorblind_cmap = LinearSegmentedColormap.from_list('colorblind_cmap', colors)
```

```python
plt.figure(figsize=(10, 8), dpi = 300)

agrietamientos=[65,65,29]
toribio=["Tacueyó", "San Francisco", "Toribío"]

# Create the pie chart with percentage labels outside
pie = plt.pie(agrietamientos, autopct='%1.1f%%', colors=colorblind_cmap(np.linspace(0, 1, len(agrietamientos))),\
              pctdistance=1.11,wedgeprops={'edgecolor': 'black'}, textprops={'fontsize': 10}, startangle = -10 )

# Set aspect ratio and equal axes to ensure a circular pie chart
plt.axis('equal')

# Move the legend outside the pie chart area
plt.legend(pie[0], [f'{r}: {c}' for r, c in zip(toribio, agrietamientos)], loc='center left', bbox_to_anchor=(1, 0.5), fontsize=14)


# Add a title
plt.title('Número de agrietamientos', fontsize = 16)
# Display the chart
plt.show()
```

```python
plt.figure(figsize=(10, 8), dpi=300)

resp = ['Número de agrietamientos', 'Número de movimientos en masa', 'Número de represamientos', 
        'Viviendas en riesgo por \n cercanía a las vías', 
        'Viviendas en riesgo por cercanía a \n movimientos en masa y agrietamientos', 
        'Viviendas en riesgo por cercanía \n al cauce de quebradas o ríos', 'Sitios de interés comunitario en riesgo']
cuentas = [159, 218, 42, 265, 311, 146, 45]
colores_pastel = ['#FFB3BA', '#FFDFBA', '#FFFFBA', '#BAFFC9', '#BAE1FF', '#D4A5A5', '#C5B9CD']



bars = plt.bar(resp, cuentas, color=colores_pastel[:len(cuentas)])

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, height, ha='center', va='bottom', fontsize=12)

plt.tight_layout()
plt.xticks(rotation=45, ha='right', fontsize = 16)
plt.ylabel('Zonas de riesgo', fontsize = 16)

plt.show()
```
